```python
#import math module

from math import sqrt

#create an empty list called math spending
math_spending = []

#add the data points
math_spending.append(('GHS', 73000))

#check my work
print (math_spending)
```

```python
#add remaining data
math_spending.extend([('LHS', 90000), ('MHS', 85000), ('RHS', 60000), ('JHS', 80000), ('CHS', 82000), ('OHS', 89000), ('IHS', 98000), ('WHS', 73000), ('AHS', 81000)])
```

```python
print(math_spending)
```

```python
#find the N

n = len(math_spending)

print(n)
```

```python
#call the amounts
amt = []

for i in math_spending:
    amt.append(i[1])
    
print(amt)
    
```

```python
s = sum(amt)

#find the mean of all curriculum budgets
mean = s//n

print(mean)
```

```python
#find the difference between the mean and each budget
diff = []

for i in amt:
    diff.append(i-mean)
    
print(diff)
```

```python
sod = []
for i in diff:
    sod.append(i*i)
    
print(sod)
```

```python
#find the mean of the squared differences

var = sum(sod)/len(sod)
```

```python
stndv = sqrt(var)
```

```python
print(stndv)
```
